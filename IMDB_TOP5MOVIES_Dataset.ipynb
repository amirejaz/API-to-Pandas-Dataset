```python
import numpy as np
import pandas as pd
import requests
```

```python
response = requests.get('https://imdb-api.com/en/API/Top250Movies/k_av7stgt7').json()
```

```python
response
```

```python
for i in response['items']:
  print(i['title'])
```

```python
title = []
IMDB_Rating = []
Rating_count = []
ranking = []
year = []

for i in response['items']:
  title.append(i['title'])
  try:
    IMDB_Rating.append(i['imDbRating'])
  except:
      IMDB_Rating.append(np.nan)
  Rating_count.append(i['imDbRatingCount'])
  ranking.append(i['rank'])
  year.append(i['year'])

d = {'title': title, 'IMDB_Rating':IMDB_Rating, 'Rating_count':Rating_count, 'ranking':ranking, 'year':year}
df = pd.DataFrame(d)

df
```

```python
df.to_csv('TOP5Movies.csv')
```
