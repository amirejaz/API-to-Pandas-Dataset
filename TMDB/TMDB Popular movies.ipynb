```python
final = pd.DataFrame()

for j in range(1, 21):
    response = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=46e9fc286c8423153158c40cacf0017f&language=en-US&page={}".format(j)).json()
    
    original_language = []
    popularity = []
    release_date = []
    title = []
    vote_average = []
    vote_count = []




    for i in response['results']:
        original_language.append(i['original_language'])
        popularity.append(i['popularity'])
        
        try:
            release_date.append(i['release_date'])
        except:
            release_date.append(np.nan())

            
            
            
        title.append(i['title'])
        vote_average.append(i['vote_average'])
        vote_count.append(i['vote_count'])



    d= {'title':title, 'vote_count':vote_count, 'vote_average':vote_average, 'release_date':release_date, 'original_language':original_language}
    df = pd.DataFrame(d)
    

    
    final = final.append(df, ignore_index=True)
    
```
